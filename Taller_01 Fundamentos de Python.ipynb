```python
# Variables

#Solicita al usuario ingresar el nombre de un producto.
nombre_producto = input("Ingresa el nombre de un producto: ")
# Pídele que ingrese el precio del producto y la cantidad vendida.
precio_producto = float(input("Ingresa el precio del producto: "))
# Convierte el precio a un número decimal (float) y la cantidad a un número entero (int).
cantidad_producto = int(input("Ingresa la cantidad vendida: "))
# Calcula el total de la venta y muestra un mensaje en pantalla con el resumen
resultado = precio_producto * cantidad_producto
print(f"El total de la venta para el producto {nombre_producto} es $ {resultado:,.2f}")
```

```python
# Estructuras de Datos (Listas, Tuplas, Diccionarios)

# Tupla
categorias_productos = ('Electrodomesticos','Mercado y alimentos','Ropa','Hogar y Limpieza','Cuidado personal','Tecnología')
print(categorias_productos)
```

```python
# Diccionario
venta = {
 
    "producto": "Zapatos",
    "categoria": 'Ropa',
    "precio": 150000, 
    "cantidad": 269,
    "total": 150000 * 269 }

# Lista
registro_ventas = []
registro_ventas.append(venta)

print(registro_ventas)
```

```python
# Estructuras de Control (Condicionales y Ciclos)
# Se agregan mas ventas
registro_ventas.append({   
    "producto": "Cereales",
    "categoria": 'Mercado y alimentos',
    "precio": 1000, 
    "cantidad":9,
    "total": 1000 * 9})

registro_ventas.append({   
    "producto": "Celulares",
    "categoria": 'Tecnología',
    "precio": 1630000, 
    "cantidad": 124,
    "total": 1630000 * 124})

registro_ventas.append({   
    "producto": "Jabones",
    "categoria": 'Hogar y Limpieza',
    "precio": 10900, 
    "cantidad": 5,
    "total": 10900 * 5})

registro_ventas.append({
    "producto": "Estufa",
    "categoria": 'Electrodomesticos',
    "precio": 2987500, 
    "cantidad": 4,
    "total": 2987500 * 4})

registro_ventas.append({
    "producto": "Crema dental",
    "categoria": 'Cuidado personal',
    "precio": 5500, 
    "cantidad": 17,
    "total": 5500 * 17})
```

```python
registro_ventas[4]
```

```python
# Condicionales y Ciclos
# Ciclo con for
for venta in registro_ventas:
    total_original = venta["total"]

    if total_original > 100000:
        total_descuento = total_original * 0.90

        print("Producto:", venta["producto"])
        print("Total original: $", (f"{total_original:,}"))
        print("Aplicar 10% de descuento")
        print("Total con descuento: $", (f"{total_descuento:,}"))

    else:
        
        print("Producto:", venta["producto"])
        print("No aplica descuento")
        print("Total:", (f"{total_original:,}"))

    print("   ")
```

```python
# Funciones y Manejo de Errores
def calcular_ingreso(precio,cantidad):
    #Intentamos multiplicar el precio por la cantidad
    try:
       ingreso = float(precio) * int(cantidad)
       return(ingreso)
    #Si en lugar de un número lleha un texto
    except:
     print("Error: Los valores deben ser númericos")
     return(0)

    
```

```python
# Valores correctos
venta_1 = calcular_ingreso(25000,20)
print("El ingreso total es de", venta_1)
```

```python
Venta_4 = calcular_ingreso(37964,85)
print("El ingreso total es de", Venta_4)
```

```python
# Valores incorrectos
venta_2 = calcular_ingreso(36852,"diez")
print("El ingreso total es de", venta_2)
```

```python
venta_3 = calcular_ingreso("tres",15)
print("El ingreso total es", venta_3)
```
